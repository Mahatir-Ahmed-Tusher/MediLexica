## **Overview of MediLexica**

MediLexica is a **medical dictionary and assistant** designed to help users understand complex medical terms and concepts. It is part of the **EarlyMed Project**, aimed at empowering patients and individuals to better comprehend medical jargon often used by healthcare professionals. By leveraging advanced language models and a curated medical knowledge base, MediLexica provides accurate, detailed, and empathetic responses to medical queries, ensuring users feel more informed and confident.

---

## Key Features of MediLexica

### 1. **Medical Knowledge Base**
   - MediLexica is built on a comprehensive medical knowledge base derived from three authoritative sources:
     1. **Webster's New World Medical Dictionary**
     2. **Dictionary of Medical Terms**
     3. **AMA Glossary of Medical Terms**
   - These resources ensure that the responses are accurate, reliable, and grounded in trusted medical literature.

### 2. **Advanced Language Model Integration**
   - MediLexica uses **Groq's Llama-3.3-70b-versatile** model, a state-of-the-art language model, to generate responses.
   - The model is fine-tuned to provide **empathetic and context-aware answers**, making it user-friendly and approachable.

### 3. **Source Citation**
   - Every response includes **citations from the Medilexica book**, ensuring transparency and credibility.
   - Users can see the exact page and content from the source material, enhancing trust and understanding.

### 4. **User-Friendly Interface**
   - The Gradio-based interface is designed to be intuitive and easy to use.
   - Features include:
     - A **prominent logo** at the top center for branding.
     - A **text input box** for entering medical queries.
     - A **response box** that displays detailed answers with source citations.

### 5. **ChromaDB for Efficient Retrieval**
   - MediLexica uses **ChromaDB**, a vector database, to store and retrieve medical information efficiently.
   - The **PDF content is split into chunks** and embedded using **sentence-transformers/all-MiniLM-L6-v2**, enabling fast and accurate retrieval of relevant information.

### 6. **Custom Prompt Template**
   - A **custom prompt template** ensures that the model responds in a consistent and helpful manner:
     - It instructs the model to act as a medical dictionary and assistant.
     - It emphasizes providing **detailed, empathetic, and example-rich responses**.
     - It ensures that **source citations** are always included.

---

## How It Works

1. **User Input**:
   - The user enters a medical-related query (e.g., "What is the treatment for severe aortic stenosis?").

2. **Query Processing**:
   - The query is processed by the **RetrievalQA chain**, which retrieves relevant information from the ChromaDB.

3. **Response Generation**:
   - The **Groq language model** generates a detailed response, including:
     - A clear explanation of the medical term or concept.
     - Examples or additional context where applicable.
     - Citations from the Medilexica book.

4. **Output Display**:
   - The response is displayed in the Gradio interface, along with the source citations.

```python
!pip install langchain_groq langchain_core langchain_community langchain-huggingface
!pip install pypdf chromadb sentence-transformers

import os
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.document_loaders import PyPDFLoader
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_groq import ChatGroq


def initial_llm():
    llm = ChatGroq(
        temperature=0,
        groq_api_key=os.getenv("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"),
        model_name="llama-3.3-70b-versatile"
    )
    return llm


def create_db():
    pdf_path = "/content/Medilexica.pdf"  # Updated to the correct PDF name

    if not os.path.exists(pdf_path):
        raise FileNotFoundError(f" PDF file not found at: {pdf_path}")

    loader = PyPDFLoader(pdf_path)
    docs = loader.load()

    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    texts = text_splitter.split_documents(docs)

    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

    db_path = "/content/chroma_db"
    vector_db = Chroma.from_documents(texts, embeddings, persist_directory=db_path)
    vector_db.persist()

    print("ChromaDB Created and Data Saved!")
    return vector_db


def setup_qachain(vector_db, llm):
    retriever = vector_db.as_retriever()

    prompt_template = """You are Medilexica, a medical dictionary and assistant. Your role is to provide accurate, detailed, and empathetic responses to medical queries. Always cite the source of your information, specifically the book "Medilexica". Include examples where applicable. If Something is not available in medilexica, you answer it from your own database.

    Context: {context}
    User: {question}
    Medilexica: """

    PROMPT = PromptTemplate(template=prompt_template, input_variables=["context", "question"])

    qa_chain = RetrievalQA.from_chain_type(
        llm=llm,
        chain_type="stuff",
        retriever=retriever,
        chain_type_kwargs={"prompt": PROMPT},
        return_source_documents=True  # Enable source citation
    )

    return qa_chain


def main():
    print("Initializing Medilexica - Your Medical Dictionary Chatbot...")

    try:
        llm = initial_llm()
        db_path = "/content/chroma_db"

        if not os.path.exists(db_path):
            print(" Creating ChromaDB...")
            vector_db = create_db()
        else:
            print(" Loading Existing ChromaDB...")
            embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
            vector_db = Chroma(persist_directory=db_path, embedding_function=embeddings)

        qa_chain = setup_qachain(vector_db, llm)

        while True:
            query = input("\nHuman: ")
            if query.lower() == "exit":
                print("Medilexica: Take care! Feel free to ask anytime. Goodbye!")
                break

            # Get response and source documents
            result = qa_chain({"query": query})
            response = result["result"]
            source_docs = result["source_documents"]

            # Format response with sources
            formatted_response = f"Medilexica: {response}\n\nSources:\n"
            for i, doc in enumerate(source_docs, 1):
                formatted_response += f"{i}. Page {doc.metadata['page'] + 1} of '/content/Medilexica.pdf': {doc.page_content[:200]}...\n"

            print(formatted_response)

    except Exception as e:
        print(f" Error: {e}")


if __name__ == "__main__":
    main()
```
